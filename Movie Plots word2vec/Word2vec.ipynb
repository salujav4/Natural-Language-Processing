```python
from gensim.test.utils import common_texts
from gensim.models import Word2Vec
from nltk.tokenize import sent_tokenize, word_tokenize
import gensim
```

```python
w2v = Word2Vec(sentences=common_texts, vector_size=100, window=5, min_count=1, workers=4)
```

```python
w2v.train([["hello", "world"]], total_examples=1, epochs=1)
```

```python
corpus_text = 'n'.join(df2[:1000]['Plot'])
df_final = []
for i in sent_tokenize(corpus_text):
    tp = []
    for j in word_tokenize(i):
        tp.append(j.lower())
    df_final.append(tp)
```

```python
model3 = gensim.models.Word2Vec(df_final, vector_size=150, alpha=0.025,window=10, min_count=2, workers=10, epochs=10)
```

```python
def most_similar(string):
    word = string
    words = model3.wv.most_similar(positive = word)
    return words
```
